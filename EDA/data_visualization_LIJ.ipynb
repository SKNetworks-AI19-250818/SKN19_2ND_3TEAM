```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('./data/netflix_user_data_cleaned.csv')
```

```python
# 구독 플랜별 사용자 수 시각화
plt.figure(figsize=(12, 7))
sns.countplot(y='Subscription Plan', data=df, order = df['Subscription Plan'].value_counts().index)
plt.title('Number of Users per Subscription Plan')
plt.xlabel('Number of Users')
plt.ylabel('Subscription Plan')
plt.show()
```

```python
# 구독 기간과 고객 만족도 간의 관계 시각화
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Subscription Length (Months)', y='Customer Satisfaction Score (1-10)', data=df, alpha=0.5)
plt.title('Subscription Length vs. Customer Satisfaction')
plt.xlabel('Subscription Length (Months)')
plt.ylabel('Customer Satisfaction Score (1-10)')
plt.show()
```

```python
# 수치형 데이터 간의 상관 관계 히트맵
numerical_cols = [
    'Subscription Length (Months)', 'Customer Satisfaction Score (1-10)', 
    'Daily Watch Time (Hours)', 'Engagement Rate (1-10)', 'Support Queries Logged', 
    'Age', 'Monthly Income ($)', 'Promotional Offers Used', 'Number of Profiles Created'
]
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

### 히트맵 상 상관관계가 크지 않기 때문에 몇몇 부분만 시각화 하여 살펴본 결과

```python
# 구독 플랜별 이탈률 시각화
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Subscription Plan', hue='Churn status')
plt.title('Churn Rate by Subscription Plan')
plt.xlabel('Subscription Plan')
plt.ylabel('Count')
plt.show()
```

```python
# 고객 만족도와 이탈 상태 간의 관계 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Churn status', y='Customer Satisfaction Score (1-10)')
plt.title('Customer Satisfaction by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Customer Satisfaction Score (1-10)')
plt.show()
```

```python
# 일일 시청 시간과 이탈 상태 간의 관계 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Churn status', y='Daily Watch Time (Hours)')
plt.title('Daily Watch Time by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Daily Watch Time (Hours)')
plt.show()
```
